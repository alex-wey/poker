```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
import statsmodels.api as sm
import statsmodels.stats.weightstats as ws
```

```python
conn = sqlite3.connect('poker_updated.db')
```

```python
df = pd.read_sql_query('select * from Hands order by player_ID, game_ID;', conn)
print(df.head())
```

```python
position_in_day = []
net_gain = []
curr_game_id = ''
curr_player_id = ''
curr_position_in_day = 0
for row in df.itertuples():
    game_id = row.game_ID[:8]
    player_id = row.player_ID
    if game_id != curr_game_id or player_id != curr_player_id:
        curr_position_in_day = 0
        curr_game_id = game_id
        curr_player_id = player_id
    position_in_day.append(curr_position_in_day)
    net_gain.append(row.net_gain)
    curr_position_in_day += 1
    
print(position_in_day[:50])
print(net_gain[:50])
```

```python
position_in_day = []
net_gain = []
curr_player_id = ''
curr_position_in_day = 0
for row in df.itertuples():
    game_id = row.game_ID[:8]
    player_id = row.player_ID
    if player_id != curr_player_id:
        curr_position_in_day = 0
        curr_player_id = player_id
    position_in_day.append(curr_position_in_day)
    net_gain.append(row.net_gain)
    curr_position_in_day += 1
    
print(position_in_day[:50])
print(net_gain[:50])
```

```python
fig,ax = plt.subplots(1, 1,figsize=(10, 10))
ax.scatter(position_in_day, net_gain, marker='.')
plt.show()
```

```python
x = sm.add_constant(position_in_day)
model = sm.OLS(net_gain, x)
```

```python

res = model.fit()
print(res.summary())
```

```python
net_gain_today = 0
after_win = []
after_loss = []
curr_game_id = ''
curr_player_id = ''
for row in df.itertuples():
    game_id = row.game_ID[:8]
    player_id = row.player_ID
    if game_id != curr_game_id or player_id != curr_player_id:
        net_gain_today = 0
        curr_game_id = game_id
        curr_player_id = player_id
        continue
    if net_gain_today > 0:
        after_win.append(row.net_gain)
    elif net_gain_today < 0:
        after_loss.append(row.net_gain)
    net_gain_today += row.net_gain


```

```python
net_gain_today = 0
after_win_players = []
after_loss_players = []
after_win = []
after_loss = []
curr_game_id = ''
curr_player_id = ''
for row in df.itertuples():
    game_id = row.game_ID[:8]
    player_id = row.player_ID
    if game_id != curr_game_id or player_id != curr_player_id:
        if player_id != curr_player_id:
            total_winrate = (np.sum(after_win) + np.sum(after_loss))\
                / (len(after_win) + len(after_loss)) if (len(after_win) + len(after_loss)) > 0 else 0
            if len(after_win) > 0 and len(after_loss) > 0:
                after_win_players.append(np.mean(after_win))
                after_loss_players.append(np.mean(after_loss))
            after_win = []
            after_loss = []
        net_gain_today = 0
        curr_game_id = game_id
        curr_player_id = player_id
        continue
    if net_gain_today > 0:
        if row.net_gain > 0:
            after_win.append(1)
        elif row.net_gain < 0:
            after_win.append(0)
    elif net_gain_today < 0:
        if row.net_gain > 0:
            after_loss.append(1)
        elif row.net_gain < 0:
            after_loss.append(0)
    net_gain_today += row.net_gain
```

```python
plt.hist(after_win_players)
plt.show()
plt.hist(after_loss_players)
plt.show()

print(np.mean(after_win_players))
print(np.mean(after_loss_players))

print(ws.ttest_ind(after_win_players, after_loss_players))
```

```python
print(np.mean(after_win))
print(np.mean(after_loss))
print(np.var(after_win))
print(np.var(after_loss))

print(ws.ttest_ind(after_win, after_loss))
```

```python
a_heights, a_bins = np.histogram(after_win, bins=2)
b_heights, b_bins = np.histogram(after_loss, bins=a_bins)

width = (a_bins[1] - a_bins[0])/3

fig, ax = plt.subplots(figsize=(15, 10))
ax.bar(a_bins[:-1], np.log(a_heights), width=width, facecolor='cornflowerblue')
ax.bar(b_bins[:-1]+width, np.log(b_heights), width=width, facecolor='seagreen')
```

```python
game_number_today = 0
net_gain_today = 0
curr_game_id = ''
curr_player_id = ''
day_length = []
net_gain = []
for row in df.itertuples():
    game_id = row.game_ID[:8]
    player_id = row.player_ID
    if game_id != curr_game_id or player_id != curr_player_id:
        if game_number_today < 1000:
            day_length.append(game_number_today)
            net_gain.append(net_gain_today)
        net_gain_today = 0
        game_number_today = 0
        curr_game_id = game_id
        curr_player_id = player_id
        continue
    net_gain_today += row.net_gain
    game_number_today += 1
```

```python
fig,ax = plt.subplots(1, 1,figsize=(18, 10))
ax.scatter(day_length, net_gain, marker='.')
plt.show()
```

```python
x = sm.add_constant(day_length)
model = sm.OLS(net_gain, x)
```

```python
res = model.fit()
print(res.summary())
```
